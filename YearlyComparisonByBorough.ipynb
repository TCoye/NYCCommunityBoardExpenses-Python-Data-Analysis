```python

df2['Agency Name'] = df2['Agency Name'].apply(lambda x: 'STATEN ISLAND COMMUNITY BOARD' if 'STATEN ISLAND COMMUNITY BOARD' in x else x)
df2 [df2['Agency Name']== 'STATEN ISLAND COMMUNITY BOARD']
```

```python
df2['Agency Name'] = df2['Agency Name'].apply(lambda x: 'BRONX COMMUNITY BOARD' if 'BRONX COMMUNITY BOARD' in x else x)
df2 [df2['Agency Name']== 'BRONX COMMUNITY BOARD']
```

```python
df2['Agency Name'] = df2['Agency Name'].apply(lambda x: 'BROOKLYN COMMUNITY BOARD' if 'BROOKLYN COMMUNITY BOARD' in x else x)
df2 [df2['Agency Name']== 'BROOKLYN COMMUNITY BOARD']
```

```python
df2['Agency Name'] = df2['Agency Name'].apply(lambda x: 'MANHATTAN COMMUNITY BOARD' if 'MANHATTAN COMMUNITY BOARD' in x else x)
df2 [df2['Agency Name']== 'MANHATTAN COMMUNITY BOARD']
```

```python
df3 =df2[(df2['Agency Name']== 'QUEENS COMMUNITY BOARD' )| (df2['Agency Name']== 'MANHATTAN COMMUNITY BOARD')| (df2['Agency Name']== 'BROOKLYN COMMUNITY BOARD')| (df2['Agency Name']== 'BRONX COMMUNITY BOARD')| (df2['Agency Name']== 'STATEN ISLAND COMMUNITY BOARD') ]
```
